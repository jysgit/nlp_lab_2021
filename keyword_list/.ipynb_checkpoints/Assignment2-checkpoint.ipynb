```python
%time
import os
import re
import string
import collections
```

```python
def tokenize(text):
    text = text.lower()
    tokens = text.translate(str.maketrans('', '', string.punctuation)).split(' ')
    
    return tokens

def calculate_frequency(tokens):
    frequency = collections.Counter(tokens)
    return frequency

def get_ngram(tokens, n=2):
    ngram = []
    for i in range(len(tokens)):
        ngram.append(' '.join(tokens[i: i+n]))
    return ngram
```

```python
file_path = os.path.join('keyword_list','BAWE.txt')
BAWE_unigram = []

# 讀每一行進來的時候要做set，取出每個文件內不重複的單字

with open(file_path, 'r') as f:
    for text in f.readlines():
        BAWE_unigram += list(set(tokenize(text.rstrip('\n'))))
```

```python
BAWE_unigram_counter = calculate_frequency(BAWE_unigram)

# 利用freq 製造紀錄word, rank的字典

BAWE_unigram_rank = {}
for i, tuples in enumerate(BAWE_unigram_counter.most_common()):
    BAWE_unigram_rank[tuples[0]] = i+1
```

```python
# Read Web1T Data
file_path = os.path.join('keyword_list', 'Web1T_unigram.txt')
Web1T_unigram_counter = {}
with open(file_path,'r') as f:
    for line in f.readlines():
        line=line.rstrip("\n").split("\t")
        Web1T_unigram_counter[line[0]] = int(line[1])
```

```python
Web1T_unigram_counter = collections.Counter(Web1T_unigram_counter)
```

```python
Web1T_unigram_Rank = {}
for i, tuples in enumerate(Web1T_unigram_counter.most_common()):
    Web1T_unigram_Rank[tuples[0]] = i+1
```

```python
rank_ratio = {}
for word, rank in BAWE_unigram_rank.items():
    if word in Web1T_unigram_Rank.keys():
        rank_ratio[word] = round(Web1T_unigram_Rank[word] / rank, 4)
    else:
        rank_ratio[word] = round(1 / rank, 4)
rank_ratio = sorted(rank_ratio.items(), key=lambda x: x[1], reverse=True)[:30]
```

```python
print('%-8s\t%-20s\t\t%-20s' %('rank', 'unigram', 'Rank Ratio'))
for i, tuples in enumerate(rank_ratio):
    print('%-8s\t%-20s\t\t%-20s' %(i+1, tuples[0], tuples[1]))
```

```python
file_path = os.path.join('keyword_list','BAWE.txt')
BAWE_bigram = []

with open(file_path, 'r') as f:
    for text in f.readlines():
        tokens = tokenize(text.rstrip('\n'))
#         把每篇文章的set放進去>>再算list 中的 count
        BAWE_bigram += list(set(get_ngram(tokens, n=2)))
```

```python
# Read Web1T Data bigram
file_path = os.path.join('keyword_list', 'Web1T_bigram.txt')
Web1T_bigram_counter = {}
with open(file_path,'r') as f:
    for line in f.readlines():
        line=line.rstrip("\n").split("\t")
        Web1T_bigram_counter[line[0]] = int(line[1])
```

```python
BAWE_bigram_count = calculate_frequency(BAWE_bigram)
```

```python
BAWE_bigram_rank = {}
i = 1
for word in BAWE_bigram_count.most_common():
    BAWE_bigram_rank[word[0]] = i
    i += 1
```

```python
Web1T_bigram_counter = collections.Counter(Web1T_bigram_counter)

Web1T_bigram_rank = {}
i = 1
for word in Web1T_bigram_counter.most_common():
    Web1T_bigram_rank[word[0]] = i
    i = i+1
```

```python
#### [TODO] calculate all rank ratios of unigrams in BAWE
bi_ans = []
for word in BAWE_bigram_rank:
    try: 
        ratio = round(Web1T_bigram_rank[word] / BAWE_bigram_rank[word], 4)
        bi_ans.append((word, ratio))
    except:
        ratio = round(1 / BAWE_bigram_rank[word], 4)
        bi_ans.append((word, ratio))
```

```python
# extract top 30 rank
bi_ans_top30 = sorted(bi_ans, key=lambda x: x[1], reverse=True)[:30]
```
